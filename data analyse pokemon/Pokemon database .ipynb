## read data in pandas

```python
# read headers 
print(df.columns)
# read each column 
```

```python
df.loc[df["Type 1"] == "Fire"].head()

```

```python
df["Total"] = df.iloc[:,4:10].sum(axis = 1)
df.head()
```

```python
cols = list(df.columns.values) 
df = df[cols[0:9] + [cols[-1]] + cols[10:12]]
df.head()
```

```python
#df.to_csv('modifiedPokemon.csv', index = False)
df.to_excel('excelModifiedPokemon.xlsx', index=False)
```

# Filter data

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')].head()

```

# Filter Total greater than 500

```python
StrongDf = df.loc[df['Total'] >= 500]
StrongDf.head()
```

# Filter Type 1 is Fire and total is greater than 500 and create a new dataframe

```python
strongFireDf = df.loc[(df['Total'] >= 500) & (df['Type 1'] == 'Fire')]
strongFireDf.head()
```

### Reset index for strong fire Pokemon dataframe

```python
strongFireDf = strongFireDf.reset_index(drop=True)
#drop = True is to remove the old index column
strongFireDf.head()
```

# Get names which contains Mega

```python
df.loc[df['Name'].str.contains('Mega')].head()
```

# Get Names without Mega

```python
df.loc[~df['Name'].str.contains('Mega')]
#      ~ means doesn't equal
```

```python
import re 
# re is regular expressions, super useful for filtering things based on textual patterns
# flags = re.I means we don't care about capitalization. 
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex = True)]
```

# Find Pokemons who begin with PI 

```python
# need to import re library 
# ^ tels begin with
# [a-z] we can have character after
# * tels 0 or more character

df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex = True)]
```

# Change Name of content in a column with condition

```python
df.loc[df['Type 1'] == 'Fire', 'Type 1'] ='Flamer' 
df.head()
```

# Change 2 columns at the time 

```python
# changing the columns Generation and Legendary if Total is greater than 500, it becomes 'Test value'
df.loc[df['Total'] > 500, ['Generation', 'Legendary']] = 'Test value'
df.head()
```

# aggregate Statistics Groupby 

```python
# mean() makes averages 
stats = df.groupby(['Type 1']).mean()
stats.head()
```

```python
#sorting values of Total descending 
stats.sort_values("Total", ascending=False)
```

# How many pokemon per type 

```python
# create new column count with the value 1
df['Count'] = 1
# group by type 1 qnd count and display count column only 
df.groupby(['Type 1']).count()['Count']
```

# make dataframe for amenities and clean the data  
### Import libraries and data

```python
import os
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',100)

all_data = pd.read_csv("Listings.csv")
all_data.head(1)
```

# Get the column amenities and delet the last character 

```python
all_data2=all_data["amenities"].str[:-1]
all_data2.head(2)
```

```python
amenities= all_data2.str.split(',', expand=True)
amenities.head(1)
```

```python
 for i in range(89):
        amenities[i]= amenities[i].str[2:-1]
amenities.head(1)
```

# Get unique values in all columns 

```python
unique_values = set()
for col in amenities:
    unique_values.update(amenities[col])
```

```python
unique_values1= pd.DataFrame(list(unique_values))
```

# Count how many times each unique value

```python
#créer table avec les valeurs uniques et le compte de ces valeurs 
count=amenities.apply(pd.value_counts)
```

```python
#insérer la colonne sum_count et additioner les valeurs 
count.insert(loc=0, column='sum_count', value=count.sum(axis=1))
```

```python
#trier les valeurs du plus grand au plus petit
count.sort_values("sum_count", axis = 0, ascending = False,
                 inplace = True, na_position ='last')
```

```python
#display the most recurent
count2=count["sum_count"]
count2
```

```python
# count2 wasn't a dataframe anymore and the names of amenities was the index 
count3=count2.to_frame().reset_index()
count3
```

# use a list of strings (count3["index"]) to replace the values in df if it contains what is in count3["index"]
## use 50 first values 
## replace with the value in count3 

```python
# use the index to clean the data, maybe use the 50 first amenities to clean the rest with .contain fonction

for y in range (50):
    count3["index"].iloc[y]
    for i in range(1,89):
        amenities.loc[amenities[i].str.contains(count3["index"].iloc[y], case=False,na=False) ,i ]= count3["index"].iloc[y]
```

# Clean some more values 

```python
#clean the data 
#Change if contain "X" replace with X 
for i in range(1,89):
    amenities.loc[amenities[i].str.contains('Paid' and 'parking', case=False,na=False) ,i ]= "Paid parking"
```

```python
for i in range(1,89):
    amenities.loc[amenities[i].str.contains('Soap', case=False,na=False) ,i ]= "Soap"
```

```python
for i in range(1,89):
    amenities.loc[amenities[i].str.contains('Free' and 'parking', case=False,na=False) ,i ]= "Free parking"
```

```python
for i in range(1,89):
    amenities.loc[amenities[i].str.contains('sound' and 'system', case=False,na=False) ,i ]= "Sound system"
```

# delete duplicate cells in a row
## replace all duplicates in a row with NaN np.nan

```python
mask = amenities.apply(pd.Series.duplicated, 1) & amenities.astype(bool)
amenities=amenities.mask(mask,np.nan)
```

```python
count=amenities.apply(pd.value_counts)
count.insert(loc=0, column='sum_count', value=count.sum(axis=1))
count.sort_values("sum_count", axis = 0, ascending = False, inplace = True, na_position ='last')
count2=count["sum_count"]
count2.head(50)
```
